SEABRON FOR VISUALIZATION

```python
import seaborn as sns
sns.set()
sns.set(style="darkgrid")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
plt.rcParams['figure.figsize']=(10,10)
```

```python
#read the dataset
df=pd.read_csv('bigmart_data.csv')
#drop the null values
df=df.dropna(how='any')
#multiply item visibility by 100 to increase size
df["Visibility_Scaled"] = df["Item_Visibility"] *100
#view top 5
df.head()

```

LINE CHART (lineplot)

```python
sns.lineplot(x="Item_Weight", y="Item_MRP", data=df[:50]);
```

BAR CHART(barplot)

```python
sns.barplot(x="Item_Type", y="Item_MRP", data= df[:5])
```

HISTOGRAM (distplot)

```python
sns.distplot(df["Item_MRP"])
```

BOXPLOTS (boxplot)


```python
sns.boxplot(df["Item_Outlet_Sales"], orient='horizontal')
```

VIOLINPLOT(violinplot)

```python
sns.violinplot(df["Item_Outlet_Sales"], orient='vertical', color='magenta')
```

SCATTERPLOT 

```python
#scatter plot
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data=df[:200], kind="scatter");
```

HUE Semantic

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", hue="Item_Type",data=df[:200]);
```

```python
# different line plots for different categories of the Outlet_Size
sns.lineplot(x="Item_Weight", y="Item_MRP",hue='Outlet_Size',data=df[:150]);
```

```python
# bubble plot
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data=df[:200], kind="scatter", size="Visibility_Scaled", hue="Visibility_Scaled");
```

```python
# subplots for each of the category of Outlet_Size
sns.relplot(x="Item_Weight", y="Item_Visibility",hue='Outlet_Size',style='Outlet_Size',col='Outlet_Size',data=df[:100]);
```
